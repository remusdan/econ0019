```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings('ignore')

df = pd.read_stata('ECON0019_empirical_project_2026.dta')

# Define regressand log(viorate)
df['log_viorate'] = np.log(df['viorate'])

# Summary statistics (mean, quantile, standard deviation)
summary = df.groupby('year')[['shallcarry','log_viorate']].describe()
display(summary)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate yearly averages - group by year first
yearly_means = df.groupby('year').agg({
    'shallcarry': 'mean',
    'log_viorate': 'mean',
    'viorate': 'mean'
}).reset_index()

# Dual-axis plot
fig, ax1 = plt.subplots(figsize=(8,8))
ax2 = ax1.twinx()
ax1.plot(yearly_means['year'], yearly_means['shallcarry'], color='blue', label='Shall Carry')
ax2.plot(yearly_means['year'], yearly_means['log_viorate'], color='red', label='Log Violence Rate')

ax1.set_xlabel('Year')
ax1.set_ylabel('Shall Carry Mean', color='blue')
ax2.set_ylabel('Log Violence Rate Mean', color='red')

min_year = yearly_means['year'].min()
max_year = yearly_means['year'].max()

step = 1
xticks = np.arange(min_year, max_year + 1, step)
ax1.set_xticks(xticks)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Yearly Averages of Shall Carry and Log Violence Rate')
plt.show()
```

```python
# Question 2 Regression Models
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_stata('ECON0019_empirical_project_2026.dta')

# MODEL 1
# Define log(viorate)
df['log_viorate'] = np.log(df['viorate'])

# Define independent and dependent variables
X = df[['shallcarry']]
X = sm.add_constant(X)
y = df['log_viorate']

# Regress y on X:
model1 = sm.OLS(y, X).fit(cov_type='cluster', 
                         cov_kwds={'groups': df['state']})
print("Model 1")
print(model1.summary())
```

```python
# MODEL II
# Define independent and dependent variables
X = df[['shallcarry','popdens','avginc','popul','pctmale','pctwhite','pctblack']]
X = sm.add_constant(X)
y = df['log_viorate']

# Regress y on X:
model2 = sm.OLS(y, X).fit(cov_type='cluster', 
                         cov_kwds={'groups': df['state']})
print("Model 2")
print(model2.summary())
```

```python
# Question 3
# Create lag variable
df = df.sort_values(['state', 'year'])
df['shallcarry_lag'] = df.groupby('state')['shallcarry'].shift(1)
df_lag = df.dropna(subset=['shallcarry_lag']).copy()

print(f"Sample size with lag: {len(df_lag)} observations")
print(f"Years covered: {df_lag['year'].min()} to {df_lag['year'].max()}")
print(f"(First year dropped for each state due to inapplicable first lag)")

# MODEL II with First Lag included
X = df_lag[['shallcarry','shallcarry_lag','popdens','avginc','popul','pctmale','pctwhite','pctblack']]
X = sm.add_constant(X) 
y = df_lag['log_viorate']

# Regress y on X:
model3 = sm.OLS(y, X).fit(cov_type='cluster', 
                         cov_kwds={'groups': df_lag['state']})
print("\nModel 2 with first lag of shallcarry\n")
print(model3.summary())
```

```python
# Question 7 Fixed Effects for log_robrate and log_murrate

import pandas as pd
from linearmodels.panel import PanelOLS
import numpy as np
df = pd.read_stata('ECON0019_empirical_project_2026.dta')

# Define log(robrate)
df['log_robrate'] = np.log(df['robrate'])

df = df.set_index(['state', 'year'])
exog_vars = ['shallcarry', 'popdens', 'avginc', 'popul', 
             'pctmale', 'pctwhite', 'pctblack']

# Estimate fixed effects model for robrate
model7a = PanelOLS(df['log_robrate'], 
                 df[exog_vars], 
                 entity_effects=True, 
                 time_effects=True).fit(cov_type='clustered', 
                                       cluster_entity=True)

print("\nlog_robrate fixed effects model\n")
print(model7a)
print(f"\nshallcarry coefficient: {model7a.params['shallcarry']:.4f}")
print(f"Clustered SE: {model7a.std_errors['shallcarry']:.4f}")
print(f"p-value: {model7a.pvalues['shallcarry']:.4f}")
```
